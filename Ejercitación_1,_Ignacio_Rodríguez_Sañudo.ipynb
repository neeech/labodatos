Hay partidos con series de penales en las que no figura el ganador y empiezan con un espacio seguido de 'win'

```python
partidos_con_penales = partidos[['penalties' in cond for cond in partidos['Win conditions']]] #Toma sólo los partidos en los que hubo definición por penales
def arreglar_condicion_de_victoria(x):
  if x[0] == ' ': #Si empieza con ' ' 
    indice = x.find('(') #Busco el resultado de la definición por penales
    if int(x[indice + 1]) > int(x[indice + 5]): #El primer número corresponde al local y el segundo al visitante
      vencedor = 'local'
    else:
      vencedor = 'visitante'
    return vencedor + x
  return x 

partidos_con_penales['Win conditions'] = partidos_con_penales['Win conditions'].apply(arreglar_condicion_de_victoria) 

partidos_con_penales['Win conditions'] = partidos_con_penales.apply(lambda x: x['Win conditions'].replace('local',x['Home Team Name']), axis=1) #Si contiene 'local', lo reemplaza por el nombre del equipo local
partidos_con_penales['Win conditions'] = partidos_con_penales.apply(lambda x: x['Win conditions'].replace('visitante',x['Away Team Name']), axis=1) #Si contiene 'visitante', lo reemplaza por el nombre del equipo visitante

partidos_con_penales['Win conditions'].value_counts()
```

```python
partidos_con_penales['Win conditions'] = partidos_con_penales['Win conditions'].apply(lambda x: x[:x.find('win')-1]) #Para cada condición de victoria, toma el nombre del país y descarta el resto
```

```python
partidos_con_penales = partidos_con_penales.rename(columns={'Win conditions':'Ganador'})
cantidad_ganados = partidos_con_penales['Ganador'].value_counts() #Cuenta la cantidad de veces que aparece cada país. Corresponde a la cantidad de victorias.
```

Los goles figuran en los eventos de cada jugador

```python
partidos = partidos.merge(jugadores, how="inner", on="MatchID") 
partidos = partidos[['MatchID', 'Stage', 'Event']]
partidos = partidos.dropna()
partidos
```

```python
partidos_con_goles = partidos[[('G' in ev or 'W' in ev or 'P' in ev) for ev in partidos['Event']]] #Toma los partidos en los que hay goles en los eventos
partidos_con_goles['Event'] = partidos_con_goles['Event'].str.split() #Separa los eventos en una lista
partidos_con_goles
```

```python
partidos_con_goles['Event'] = partidos_con_goles['Event'].apply(lambda x: [ev for ev in x if ('G' in ev or 'W' in ev or ('P' in ev and not 'MP' in ev))]) #Toma sólo los eventos que son goles
partidos_con_goles
```

```python
partidos_con_goles['Event'] = partidos_con_goles['Event'].apply(lambda x: [int(''.join([n for n in ev if n.isdigit()])) for ev in x]) #De cada gol, toma la parte numérica -> minuto del gol.
partidos_con_goles_separados = partidos_con_goles.explode('Event') #Separa cada gol en filas distintas
partidos_con_goles_separados
```

```python
partidos_con_goles_separados = partidos_con_goles_separados.dropna()
partidos_con_goles_separados['Event'].value_counts()
```
